<b>2. From the Pillow module, how do you get the RGBA value of any images?<b>

```python
from PIL import Image

img = Image.open("download.jpg","r")
rgba = img.convert("RGBA")
rgba_val = rgba.getdata()
print(list(rgba_val))
```

<b>3. What is a box tuple, and how does it work?<b>

Ans: The rectangle(xy, fill, outline) method draws a rectangle. The xy argument is a box tuple of the form (left, top, right, bottom). The left and top values specify the x- and y-coordinates of the upper-left corner of the rectangle, while right and bottom specify the lower-right corner. The optional fill argument is the color that will fill the inside of the rectangle. The optional outline argument is the color of the rectangle’s outline.
